```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("swe.csv").set_index("time")
```

```python
n = len(df.columns)
l = 100.0  # from example/swe.rs
x = np.arange(0, l, l/n)
```

```python
X, T = np.meshgrid(df.index, x)
plt.pcolormesh(T, X, df.T)
plt.colorbar()
plt.xlabel("space")
plt.ylabel("time")
plt.savefig("swe.png", bbox_inches="tight", transparent=False)
```
